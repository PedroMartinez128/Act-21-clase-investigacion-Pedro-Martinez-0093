```python
class Investigador:
    def __init__(self, nombre, especialidad, edad):
        self.nombre = nombre
        self.especialidad = especialidad
        self.edad = edad

    def investigar(self):
        print(f"{self.nombre} está realizando una investigación profunda en el área de {self.especialidad}.")

    def analizar(self):
        print(f"{self.nombre} está analizando los datos recolectados. ¡Parece que hay hallazgos importantes!")

    def difusion_de_resultados(self):
        print(f"Publicación oficial: El investigador de {self.edad} años, {self.nombre}, ha compartido sus resultados con la comunidad científica.")

print("--- Datos del Investigador 1 ---")
nom1 = input("Nombre: ")
esp1 = input("Especialidad: ")
ed1 = input("Edad: ")
investigador1 = Investigador(nom1, esp1, ed1)

print("-" * 30)

print("--- Datos del Investigador 2 ---")
nom2 = input("Nombre: ")
esp2 = input("Especialidad: ")
ed2 = input("Edad: ")
investigador2 = Investigador(nom2, esp2, ed2)

print("\n--- Actividades del Investigador 1 ---")
investigador1.investigar()
investigador1.analizar()
investigador1.difusion_de_resultados()

print("\n--- Actividades del Investigador 2 ---")
investigador2.investigar()
investigador2.analizar()
investigador2.difusion_de_resultados()

print("\n===== programa terminado =====")
print("--- software realizado por Pedro Martinez NL = 34 ---")
```
